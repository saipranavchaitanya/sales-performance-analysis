```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
df.head()
```

```python
df = df.dropna()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df = df[df['Sales'] > 0]
```

```python
df = df.drop_duplicates()
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.month_name()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()

print("Total Sales:", round(total_sales,2))
print("Total Profit:", round(total_profit,2))
print("Total Orders:", total_orders)
```

```python
monthly_sales = df.groupby(['Year','Month'])['Sales'].sum().reset_index()

plt.figure()
plt.plot(monthly_sales['Month'], monthly_sales['Sales'])
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
category_analysis = df.groupby('Category')[['Sales','Profit']].sum().reset_index()
category_analysis
```

```python
plt.figure()
plt.bar(category_analysis['Category'], category_analysis['Sales'])
plt.title('Sales by Category')
plt.show()
```

```python
region_analysis = df.groupby('Region')[['Sales','Profit']].sum().reset_index()
region_analysis
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.show()
```

```python
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100

profit_margin = df.groupby('Category')['Profit_Margin'].mean().reset_index()
profit_margin
```
